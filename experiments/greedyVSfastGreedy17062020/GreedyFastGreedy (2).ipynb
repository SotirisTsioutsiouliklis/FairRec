```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the source nodes.
sourceNodes = []
with open('randomSourceNodes.txt', 'r') as fileOne:
    for line in fileOne:
        sourceNodes.append(int(line))
```

```python
sourceNodes
```

```python
# Plot the Fairness for each Source Node.
plt.figure()
for node in range(9):
    greedyRed = []
    fastGreedyRed = []
    with open('%dRedPagerankGreedy.txt' %sourceNodes[node], 'r') as fileOne:
        for line in fileOne:
            greedyRed.append(float(line))
    with open('%dRedPagerankFastGreedy.txt' %sourceNodes[node], 'r') as fileOne:
        for line in fileOne:
            fastGreedyRed.append(float(line))
    
    plt.subplot(330 + node + 1)
    plt.plot(np.arange(len(greedyRed)),np.array(greedyRed), '-.b')
    plt.plot(np.arange(len(greedyRed)),np.array(fastGreedyRed), '-.r')
```
