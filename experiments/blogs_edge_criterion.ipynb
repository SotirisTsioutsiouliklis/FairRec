```python
# Import libraries.
%matplotlib notebook
import pandas as pd
```

```python
# Load data.
grd_20_50 = pd.read_csv('out_greedy_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Greedy'])
rnd = pd.read_csv('out_random_edges_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Random'])
one = pd.read_csv('out_one_edges_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Criterion_product'])
two = pd.read_csv('out_two_edges_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Criterion_sum'])
#one.head()
```

```python
grd_40_25 = pd.read_csv('out_greedy_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Greedy2'])
```

```python
grd_10_100 = pd.read_csv('out_greedy_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Greedy3'])
```

```python
grd_50_20 = pd.read_csv('out_greedy_all.txt', sep='\t', usecols=[0, 1], header=0, names=['Edge', 'Greedy4'])
```

```python
# Join data.
df = pd.merge(grd_20_50, rnd, how='right', on='Edge')
df = df.merge(one, how='right', on='Edge')
df = df.merge(two, how='right', on='Edge')
df = df.merge(grd_40_25, how='right', on='Edge')
df = df.merge(grd_10_100, how='right', on='Edge')
df = df.merge(grd_50_20, how='right', on='Edge')
df.head()
```

```python
# Plot.
df.plot(x =0, y=[1,2,3,4,5,6,7])
```
